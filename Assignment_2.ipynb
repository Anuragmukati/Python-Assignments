### 5. (a) Generate 1000 random numbers from [0,1) and store the first 500 in vector x and remaining in vector y. (b) Find z = x ∗ (x + y). Plot scatter plot for representing z. Set axis labels and figure title.

```python
import numpy as np

random_numbers = np.random.rand(1000)
x = random_numbers[:500]
y = random_numbers[500:]
```

```python
import matplotlib.pyplot as plt

z = x * (x + y)

plt.scatter(range(len(z)), z)
plt.xlabel('Index')
plt.ylabel('z')
plt.title('Scatter Plot of z')
plt.show()
```

### 6. x=[1,-2,4], y=[-3,2,4]. Plot the hyperplane z = 2x + 3y + 6. Set label and title.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, -2, 4])
y = np.array([-3, 2, 4])
z = 2 * x + 3 * y + 6

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='red')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Hyperplane z = 2x + 3y + 6')

plt.show()
```

### 7. Plot a figure using the following information.(Figures should have axis labels and title) (a) f : [−3, 3] → R where f(x) = cosx + 1/ (1 + exp − (2x)) (b) f : [−16, 10] → R where f(x) = 2 + 2x/(5sinx − 6)  Draw the tangent line to f at x=-6. (c) f : [0, 2] × [0, 1] → R where f(x) = ∥x∥^2.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y = np.cos(x) + 1 / (1 + np.exp(-2*x))

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = cos(x) + 1 / (1 + exp(-2x))')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-16, 10, 100)
y = 2 + 2*x**5*np.sin(x) - 6

tangent_x = -6
tangent_y = 2 + 2*tangent_x**5*np.sin(tangent_x) - 6
slope = 10*tangent_x**4*np.sin(tangent_x) + 2*x**5*np.cos(tangent_x)
tangent_line = slope*(x - tangent_x) + tangent_y

plt.plot(x, y, label='f(x) = 2 + 2x^5*sin(x) - 6')
plt.plot(x, tangent_line, label='Tangent at x = -6')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = 2 + 2x^5*sin(x) - 6 with Tangent at x = -6')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x)')
ax.set_title('Plot of f(x) = ||x||^2')

plt.show()
```
