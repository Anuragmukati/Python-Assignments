### 1. Plot normal distribution with mean 0 and standard deviation 1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x-axis
x = np.linspace(-5, 5, 1000)

# Calculate the corresponding y-values using the normal distribution formula
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Plot the normal distribution
plt.plot(x, y)

# Set labels and title
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution (mean=0, std=1)')

# Display the plot
plt.show()
```

### 2. Plot exp(x), exp(âˆ’x), logx(x > 0) in the same figure by giving different shades to each plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x-axis
x = np.linspace(0.001, 5, 1000)  # Start from a small value to avoid log(0)

# Calculate the corresponding y-values for each function
exp_x = np.exp(x)
exp_neg_x = np.exp(-x)
log_x = np.log(x)

# Plot the functions with different shades
plt.plot(x, exp_x, label='exp(x)', color='blue', alpha=0.3)
plt.plot(x, exp_neg_x, label='exp(-x)', color='red', alpha=0.3)
plt.plot(x, log_x, label='log(x)', color='green', alpha=0.3)

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of exp(x), exp(-x), log(x)')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

### 3. (a) Generate 500 samples from Bernoulli distribution having probability of success Ï† = 0.25. Plot the probability mass function using the selected data.(b) Generate 2000 samples from Normal distribution having mean 100 and standard deviation 5. Plot the histogram and probability density function in the same figure using the selected data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 500 samples from Bernoulli distribution
p = 0.25
samples = np.random.binomial(1, p, size=500)

# Calculate the probability mass function (PMF)
pmf = np.bincount(samples) / len(samples)

# Plot the PMF
plt.bar([0, 1], pmf)
plt.xticks([0, 1], ['Failure', 'Success'])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of Bernoulli Distribution (Ï†=0.25)')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 2000 samples from Normal distribution
mean = 100
std_dev = 5
samples = np.random.normal(mean, std_dev, size=2000)

# Plot the histogram
plt.hist(samples, bins=30, density=True, alpha=0.7, label='Histogram')

# Calculate the probability density function (PDF)
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Plot the PDF
plt.plot(x, pdf, 'r', label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Histogram and Probability Density Function of Normal Distribution (Î¼=100, Ïƒ=5)')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

### 4. (a) Generate 200 data points that satisfy the equation 3x + 4y âˆ’ 3 = 0. Plot the hyperplane and mark the selected points. (b) Generate 1000 data points that satisfy the equation 5x âˆ’ 2y âˆ’ 6z = 0. Plot the hyperplane and mark the selected points.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x and y values
x = np.random.uniform(-10, 10, size=200)
y = (3 - 3*x) / 4

# Plot the hyperplane
plt.plot(x, y, 'b-', label='Hyperplane')

# Plot the selected points
plt.scatter(x, y, c='r', label='Selected Points')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperplane: 3x + 4y - 3 = 0')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x, y values
x = np.random.uniform(-10, 10, size=200)
y = np.random.uniform(-10, 10, size=200)

# Calculate z values that satisfy the equation 5x - 2y - 6z = 0
z = (5*x - 2*y) / -6

# Plot the hyperplane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(x, y, z, linewidth=0, antialiased=False)
# Plot the selected points
ax.scatter(x, y, z, c='r', label='Selected Points')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Hyperplane: 5x - 2y - 6z = 0')

# Add legend
ax.legend()

# Display the plot
plt.show()
```

### 5. Write a function module for the following tasks: (a) To determine the probability mass function of Bernoulli and Binomial distribution. (b) To determine the probability density function of Normal distribution.

```python
import math

def bernoulli_pmf(p, k):
    """Calculate the probability mass function (PMF) of a Bernoulli distribution."""
    if k == 0:
        return 1 - p
    elif k == 1:
        return p
    else:
        return 0

def binomial_pmf(n, p, k):
    """Calculate the probability mass function (PMF) of a Binomial distribution."""
    coef = math.comb(n, k)
    return coef * (p ** k) * ((1 - p) ** (n - k))

def normal_pdf(mu, sigma, x):
    """Calculate the probability density function (PDF) of a Normal distribution."""
    coefficient = 1 / (sigma * math.sqrt(2 * math.pi))
    exponent = -((x - mu) ** 2) / (2 * (sigma ** 2))
    return coefficient * math.exp(exponent)
```

### 6. Draw a figure that consists of the following subfigures: (a) Circle 1: x^2 + y^2 = 3 (b) Circle 2: x^2 + y^2 = 2 (c) Red colour between Circle 1 and Circle 2 (d) Green colour inside Circle 2.

```python
import numpy as np
import matplotlib.pyplot as plt

# Circle 1
theta = np.linspace(0, 2*np.pi, 100)
x1 = np.sqrt(3) * np.cos(theta)
y1 = np.sqrt(3) * np.sin(theta)

# Circle 2
x2 = np.sqrt(2) * np.cos(theta)
y2 = np.sqrt(2) * np.sin(theta)

# Create the figure and subplots
fig, axs = plt.subplots(1,2)

# Plot Circle 1
axs[0].plot(x1, y1, 'b')
axs[0].set_aspect('equal')
axs[0].set_title('Circle 1')

# Plot Circle 2
axs[1].plot(x2, y2, 'r')
axs[1].set_aspect('equal')
axs[1].set_title('Circle 2')

# Adjust the spacing between subplots
fig.tight_layout()

# Display the figure
plt.show()
```

```python

fig, axs = plt.subplots()

# Plot Circle 1
axs.plot(x1, y1, 'b', label='Circle 1: x^2 + y^2 = 3')
axs.fill_between(x1, y1, color='r')

# Plot Circle 2
axs.plot(x2, y2, 'y', label='Circle 2: x^2 + y^2 = 2')
axs.fill_between(x2, y2, color='g')

axs.set_aspect('equal')
axs.set_xlim(-np.sqrt(3), np.sqrt(3))
axs.set_ylim(-np.sqrt(3), np.sqrt(3))
axs.set_xlabel('x')
axs.set_ylabel('y')
axs.set_title('Circles and Colored Regions')
axs.legend()
```

### 7. Draw a figure that consists of the following subfigures: (a) Line 1: 6x âˆ’ 9y + 2 = 0 (b) Line 2: 6xâˆ’9y+2 = 1 (c) Line 3: 6xâˆ’9y+2 = âˆ’1 (d) A segment that connects 6x âˆ’ 9y + 2 = 1 and 6x âˆ’ 9y + 2 = âˆ’1. (e) Blue colour above 6x âˆ’ 9y + 2 = 1 and yellow colour below 6x âˆ’ 9y + 2 = âˆ’1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Line 1: 6x - 9y + 2 = 0
x = np.linspace(-10, 10, 100)
y = (6*x + 2) / 9

# Line 2: 6x - 9y + 2 = 1
y2 = (6*x + 1) / 9

# Line 3: 6x - 9y + 2 = -1
y3 = (6*x - 1) / 9

# Create the figure and subplots
fig, axs = plt.subplots(2, 2)

# Plot Line 1
axs[0, 0].plot(x, y, 'r')
axs[0, 0].set_xlim(-10, 10)
axs[0, 0].set_ylim(-10, 10)
axs[0, 0].set_aspect('equal')
axs[0, 0].set_title('Line 1: 6x - 9y + 2 = 0')

# Plot Line 2
axs[0, 1].plot(x, y2, 'b')
axs[0, 1].set_xlim(-10, 10)
axs[0, 1].set_ylim(-10, 10)
axs[0, 1].set_aspect('equal')
axs[0, 1].set_title('Line 2: 6x - 9y + 2 = 1')

# Plot Line 3
axs[1, 0].plot(x, y3, 'g')
axs[1, 0].set_xlim(-10, 10)
axs[1, 0].set_ylim(-10, 10)
axs[1, 0].set_aspect('equal')
axs[1, 0].set_title('Line 3: 6x - 9y + 2 = -1')

# Plot the segment connecting Line 2 and Line 3
axs[1, 1].plot(x, y2, 'b')
axs[1, 1].plot(x, y3, 'g')
axs[1, 1].fill_between(x, y2, y3, where=(y2 >= y3), color='yellow')
axs[1, 1].fill_between(x, y2, y3, where=(y2 < y3), color='blue')
axs[1, 1].set_xlim(-10, 10)
axs[1, 1].set_ylim(-10, 10)
axs[1, 1].set_aspect('equal')
axs[1, 1].set_title('Colored Regions')

# Adjust the spacing between subplots
fig.tight_layout()

# Display the figure
plt.show()
```
